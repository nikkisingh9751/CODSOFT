```python
from tkinter import *
import pickle

# Creating function to add a task
def add_task():
    task = task_entry.get()
    if task:
        todo_list.append(task)
        list_box.insert(END, task)
        task_entry.delete(0, END)

# Creating function to remove a task
def remove_task():
    selected_task = list_box.curselection()
    if selected_task:
        index = selected_task[0]
        list_box.delete(index)
        del todo_list[index]

# Creating function to mark a task as done
def mark_task():
    selected_task = list_box.curselection()
    if selected_task:
        index = selected_task[0]
        task = list_box.get(index)
        if task.startswith("✔"):
            list_box.itemconfig(index, fg="black")
            list_box.delete(index)
            list_box.insert(index, task[1:])
        else:
            list_box.itemconfig(index, fg="green")
            list_box.delete(index)
            list_box.insert(index, "✔" + task)

# Creating function to save the tasks
def save_tasks():
    with open('todo_list.pkl', 'wb') as f:
        pickle.dump(todo_list, f)

# Creating function to load the tasks
def load_tasks():
    try:
        with open('todo_list.pkl', 'rb') as f:
            todo_list.clear()
            todo_list.extend(pickle.load(f))
            list_box.delete(0, END)
            for task in todo_list:
                list_box.insert(END, task)
    except FileNotFoundError:
        todo_list.clear()

root = Tk()
root.title("To-Do List")
root.geometry("720x480")
root.resizable(False, False)
root.config(bg="lightGrey")

# Heading
title = Label(root, text="To-Do List", font=("Consolas", 18), bg="Grey", fg="#fff")
title.pack()

# Entry
text = StringVar()
task_entry = Entry(root, width=34, textvariable=text, font=("Consolas", 12))
task_entry.pack()

# Buttons
add_button = Button(root, text="Add", width=5, font=("Consolas", 12), bg="Grey", fg="#fff", command=add_task)
add_button.place(x=205, y=110)

remove_button = Button(root, text="Remove", width=6, font=("Consolas", 12), bg="Grey", fg="#fff", command=remove_task)
remove_button.place(x=450, y=110)

mark_button = Button(root, text="Mark as Done", width=12, font=("Consolas", 12), bg="Grey", fg="#fff", command=mark_task)
mark_button.place(x=300, y=130)

save_button = Button(root, text="Save", width=5, font=("Consolas", 12), bg="Grey", fg="#fff", command=save_tasks)
save_button.place(x=205, y=150)

load_button = Button(root, text="Load", width=6, font=("Consolas", 12), bg="Grey", fg="#fff", command=load_tasks)
load_button.place(x=450, y=150)

# Listbox
todo_list = []
list_box = Listbox(root, height=15, width=45, font=("Consolas", 12))
list_box.place(x=150, y=200)

root.mainloop()
```
