```python
# Wap to make a calculator using 
from tkinter import*

def Click (numbers):
    global operator
    operator=operator+ str(numbers)
    text_Input.set(operator)
    
def Clear():
    global operator
    operator=""
    text_Input.set("")
    
def Equals():
    global operator
    sumup=str(eval(operator))
    text_Input.set(sumup)
    operator=""
    
cal = Tk()
cal.title("Calculator")
operator=""
text_Input =StringVar()

#--------------------------------Display Screen-----------------------------------

txtDisplay = Entry (cal, font=('arial', 20, 'bold'), textvariable=text_Input, bd=30, insertwidth=4,
                    bg="light grey", justify='right').grid(columnspan=4)

#-----------------------------------row 1---------------------------------------

btn7=Button (cal, padx=16,pady=16, bd=8, fg="black", font=('arial', 20, 'bold'),
             text="7",bg="light grey",command=lambda:Click(7)).grid(row=1,column=0)

btn8=Button(cal,padx=16, pady=16, bd=8, fg="black", font=('arial', 20, 'bold'),
                text="8", bg="light grey",command=lambda:Click(8)).grid(row=1,column=1)

btn9=Button(cal,padx=16, pady=16, bd=8, fg="black", font=('arial', 20, 'bold'),
             text="9", bg="light grey",command=lambda:Click(9)).grid(row=1,column=2)

Addition=Button(cal,padx=16, pady=16,bd=8, fg="black", font=('arial', 20, 'bold'),
            text="+", bg="light grey",command=lambda:Click("+")).grid(row=1,column=3)

#-----------------------------------row 2---------------------------------------

btn4=Button (cal, padx=16, pady=16, bd=8, fg="black", font=('arial', 20, 'bold'),
             text="4", bg="light grey",command=lambda:Click(4)).grid(row=2,column=0)

btn5=Button(cal,padx=16, pady=16, bd=8, fg="black", font=('arial', 20, 'bold'),
                text="5", bg="light grey",command=lambda:Click(5)).grid(row=2,column=1)

btn6=Button(cal,padx=16, pady=16, bd=8, fg="black", font=('arial', 20, 'bold'),
             text="6", bg="light grey",command=lambda:Click(6)).grid(row=2,column=2)

Subtraction=Button(cal,padx=16, pady=16, bd=8, fg="black", font=('arial', 20, 'bold'),
            text="-", bg="light grey",command=lambda:Click("-")).grid(row=2,column=3)

#-----------------------------------row 3---------------------------------------

btn1=Button (cal, padx=16, pady=16, bd=8, fg="black", font=('arial', 20, 'bold'),
             text="1", bg="light grey",command=lambda:Click(1)).grid(row=3,column=0)

btn2=Button(cal,padx=16, pady=16, bd=8, fg="black", font=('arial', 20, 'bold'),
                text="2", bg="light grey",command=lambda:Click(2)).grid(row=3,column=1)

btn3=Button(cal,padx=16, pady=16, bd=8, fg="black", font=('arial', 20, 'bold'),
             text="3", bg="light grey",command=lambda:Click(3)).grid(row=3,column=2)

Multiply=Button(cal,padx=16, pady=16, bd=8, fg="black", font=('arial', 20, 'bold'),
            text="*", bg="light grey",command=lambda:Click("*")).grid(row=3,column=3)

#-----------------------------------row 4---------------------------------------

btn0=Button (cal, padx=16, pady=16, bd=8, fg="black", font=('arial', 20, 'bold'),
             text="0", bg="light grey",command=lambda:Click(0)).grid(row=4,column=0)

btnClear=Button(cal,padx=16, pady=16, bd=8, fg="black", font=('arial', 20, 'bold'),
                text="C", bg="light grey",command=Clear).grid(row=4,column=1)

btnEquals=Button(cal,padx=16, pady=16, bd=8, fg="black", font=('arial', 20, 'bold'),
             text="=", bg="light grey",command=Equals).grid(row=4,column=2)

Division=Button(cal,padx=16, pady=16,  bd=8, fg="black", font=('arial', 20, 'bold'),
            text="/", bg="light grey",command=lambda:Click("/")).grid(row=4,column=3)

#-----------------------------------Program End---------------------------------------

cal.mainloop ()
```
